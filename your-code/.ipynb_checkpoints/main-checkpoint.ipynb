#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
plt.plot(x,y,"r")
plt.show()

plt.plot(x,z,"g")
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
(fig , ax_lst) = plt.subplots(nrows=1, ncols=2,sharex=True)
ax_lst[0].plot(x,y)
ax_lst[1].plot(x,z)
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
# Method 2
(fig , ax_lst) = plt.subplots(nrows=1, ncols=2,sharex=True,figsize=(8,6))
ax_lst[0].plot(x,y)
ax_lst[1].plot(x,z)

ax_lst[0].set_xlabel('x', fontsize=8, fontweight='demibold')
ax_lst[0].set_ylabel('y', fontsize=8, fontweight='heavy')

ax_lst[1].set_xlabel('x', fontsize=8, fontweight='demibold')
ax_lst[1].set_ylabel('z', fontsize=8, fontweight='heavy')

ax_lst[0].set_title('Plot 1', fontsize=8, fontstyle = 'italic')
ax_lst[1].set_title('Plot 2', fontsize=8, fontstyle = 'italic')

fig.suptitle('Plot Tests', fontsize=10, fontweight='extra bold')
```

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
import numpy
from math import exp
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)
#y2 = np.array(list(map(exp,x)))
```

```python
plt.plot(x,y1,"r")
plt.plot(x,y2,"g")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y1, label = 'X^2')
ax.plot(x,y2, label = 'EXP(X)')

ax.set_xlabel('x', fontsize=8, fontweight='demibold')
ax.set_ylabel('y', fontsize=8, fontweight='heavy')

ax.set_xscale('log')
ax.set_yscale('log')
```

```python
fig,ax = plt.subplots()
ax.plot(x,y1, label = 'X^2')
ax.plot(x,y2, label = 'EXP(X)')

ax.set_xlabel('x', fontsize=8, fontweight='demibold')
ax.set_ylabel('y', fontsize=8, fontweight='heavy')

ax.set_xscale('log')
ax.set_yscale('log')

ax.legend()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_months = fitbit2[['Steps','Months_encoded']].groupby('Months_encoded').median()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_week = fitbit2[['Steps','Work_or_Weekend']].groupby('Work_or_Weekend').median()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))


plt.sca(ax[0])
plt.plot(steps_months)
plt.xticks()
plt.ylabel('steps')
plt.xlabel('months')

plt.sca(ax[1])
plt.bar(x = steps_week['Steps'].index, height = steps_week['Steps'])
plt.xticks([0 , 1])
plt.xlabel('days')


fig.suptitle('Average number of steps')
fig.savefig('Steps.png')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

fig, ax_list = plt.subplots(1,3, figsize=(15,8))
columns = ['Minutes Sedentary', 'Minutes Lightly Active', 'Minutes Very Active']

for axes,column in zip(ax_list, columns):
    plt.sca(axes)
    plt.scatter(fitbit2['Steps'],fitbit2[column])
    plt.ylabel('steps')
    plt.xlabel(column)

fig.suptitle('Steps vs Minutes')
```
